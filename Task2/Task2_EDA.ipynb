```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_superstore.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
# Insight:
# Technology category has highest sales
# Some categories have high sales but low profit
```

📊 Insights from Data Analysis
Technology category has the highest sales among all categories.
Profit does not always follow sales; some categories have high sales but lower profit.
A few top products contribute significantly to total revenue.
There is an uneven distribution of sales across products.
Business can focus more on high-profit categories for better growth.
